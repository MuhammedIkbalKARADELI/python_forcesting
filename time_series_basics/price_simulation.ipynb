```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
T = 1000 # Number of time steps
P0 = 10  # Initial price
mu = 0.001 # Drift

last_p = np.log(P0) # Last log price
log_returns = np.zeros(T)
prices = np.zeros(T)
```

```python
for t in range(T):
    # Sample a log return
    r = 0.01 * np.random.randn()

    # Compute the new log price
    p = last_p + mu + r

    # Stor the return and price
    log_returns[t] = r + mu
    prices[t] = np.exp(p)

    # assigned the last_p
    last_p = p

plt.figure(figsize=(20,8))
plt.plot(prices)
plt.title("Simulated Prices")
plt.show()
```
